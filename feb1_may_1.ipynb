```python
positiveCaseList = []
for i in range(1):
    SEIR_network = SEIRNetwork()
    simulator = Simulator(SEIR_network)
    simulator.generate_simulation(T = 120)
    # +++++ Feb 1st ~ May 1st +++++
    simulator.set_imported_case(time=1, average_introductions_per_day=0.1)
    simulator.set_trace_together(
        time=49, 
        tracing_lag=1, 
        tracing_compliance_rate=0.95
        )
    simulator.set_social_gathering_limit(
        time=57, 
        group_size=10
        )
    simulator.set_imported_case(time=25, average_introductions_per_day=0.4)
    simulator.set_imported_case(time=40, average_introductions_per_day=1.0)
    simulator.set_social_distancing(
        time=68, 
        global_rate=0.001
        )
    simulator.simulation.set_seir_network_params(
        time=68,
        G=simulator.SEIRModel.get_network("household")
        )
    simulator.run()
    positiveCaseList.append(simulator.simulation.numPosTseries)
    print(f"Trial({i}) is done!")

from collections import defaultdict
from statistics import median


avgPosCase = defaultdict(list)
for iteration in positiveCaseList:
    for time_float,posCase in iteration.items():
        time = int(time_float)
        avgPosCase[time].append(posCase)
for key in avgPosCase:
    avgPosCase[key] = median(avgPosCase[key])
avgPosCase

import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

figure(figsize=(15,5))
plt.bar(avgPosCase.keys(),avgPosCase.values())
```
